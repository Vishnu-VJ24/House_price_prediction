**HOUSE PRICE PREDICTION USING RANDOM FOREST ALGORITHM**

BASIC PREPROCESSINGS

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("/content/drive/MyDrive/House price prediction/Data.csv")
df
```

```python
df["Price"]=df["price"]
df.drop(["price","builder"],axis=1,inplace=True)
df.dropna(inplace=True)
df
```

```python
location_stats=df.groupby("location")["location"].count().sort_values(ascending=False)
location_stats
```

```python
location_lessthan10=location_stats[location_stats<10]
df.location=df.location.apply(lambda x: "others" if x in location_lessthan10 else x)
```

```python
dummies=pd.get_dummies(df.location)
```

```python
df1=pd.concat([df,dummies.drop("others",axis=1)],axis=1)
df1
```

```python
df1=df1.drop(["location","status"],axis=1)
```

finally preprocessings over...!

**MODEL BUILDING...***

```python
x= df1.drop(["Price"],axis=1)
y=df1["Price"]
```

```python
y.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(x_train, y_train)
```

```python
predictions = rf.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predictions)
print('Mean Absolute Error:', mae)
```

```python
rf.score(x_train,y_train)
```

**PREDICTIONS:**

```python
def predict_price(area,bhk,bathroom,age,location):
  loc_index=np.where(x.columns==location)[0][0]
  a=np.zeros(len(x.columns))
  a[0]=area
  a[1]=bhk
  a[2]=bathroom
  a[3]=age
  if loc_index >= 0:
    a[loc_index]=1

  price=rf.predict([a])[0]
  print("Price in Lakhs is",price)
  return
```

```python
predict_price(1000,2,2,1,'Vanagaram')
```

```python
predict_price(600,12,22,9,'Vanagaram')
```

```python
predict_price(1000,2,2,1,'Veppampattu')
```
